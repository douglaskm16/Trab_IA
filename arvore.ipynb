```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import (
    DecisionTreeClassifier,
    plot_tree
)

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    ConfusionMatrixDisplay
)

# Carregar o dataset
url = "https://docs.google.com/uc?export=download&id=1wsNFh-P7xpWRQ8vdZ4mTrQCfNVYVJfB3"

df = pd.read_csv(url)

print("Formato original:", df.shape)

# Remover duplicatas
duplicatas = df.duplicated().sum()

print("Duplicatas encontradas:", duplicatas)

df.drop_duplicates(inplace=True)

print("Formato após remoção:", df.shape)
```

```python
# Separar atributos e alvo
X = df.drop("target", axis=1)
y = df["target"]

# Dividir treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

#  Criar a árvore
modelo_arvore = DecisionTreeClassifier(
    max_depth=4,
    random_state=42
)

# Treinar o modelo
modelo_arvore.fit(X_train, y_train)

# Fazer previsões
y_pred = modelo_arvore.predict(X_test)

# Avaliar o modelo
print("=== Árvore de Decisão ===")

print(f"Acurácia: {accuracy_score(y_test, y_pred):.2%}")

print("\nRelatório de Classificação:")

print(classification_report(
    y_test,
    y_pred,
    target_names=[
        "Sem Doença",
        "Com Doença"
    ]
))
```

```python
# Matriz de Confusão
ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred,
    display_labels=[
        "Sem Doença",
        "Com Doença"
    ]
)

plt.title("Matriz de Confusão - Árvore de Decisão")

plt.show()
```

```python
#  Visualizar a árvore
plt.figure(figsize=(20, 10))

plot_tree(
    modelo_arvore,
    feature_names=X.columns,
    class_names=[
        "Sem Doença",
        "Com Doença"
    ],
    filled=True,
    rounded=True,
    fontsize=8
)

plt.title("Árvore de Decisão - Heart Disease Dataset")

plt.show()
```
